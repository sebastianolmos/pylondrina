# Ejemplos de uso de los wrappers para generar datasets sinteticos

```python
from pathlib import Path
import sys

REPO_ROOT = Path("../..").resolve()
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

DATA_PATH = REPO_ROOT / "data" / "synthetic" 
```

```python
from scripts.synthetic_data.gen_wrappers import (
    make_happy_path_minimal,
    make_happy_path_rich,
    make_h3_derivable,
    make_tier2_valid,
    make_tier2_mixed_invalid,
    make_multistage_valid,
    make_extended_domains,
    make_missing_required,
    make_duplicate_movement_id,
)
```

### 1) Happy path mínimo

```python
df = make_happy_path_minimal(filas=6, seed=42)
df.head()
```

Con override: agregamos dos campos base para que no sea tan “vacío”

```python
df = make_happy_path_minimal(
    filas=8,
    seed=42,
    base_fields=["mode", "purpose"],
)

display(df.head())
```

### 2) Happy path rico

```python
df = make_happy_path_rich(filas=12, seed=42)
df.head()
```

Con un pequeño override:

```python
df = make_happy_path_rich(
    filas=12,
    seed=42,
    extra_columns=["household_id", "travel_time_min", "fare_amount"],
)
display(df.head())
```

Con override: Elegimos extras concretos. Sirviendo para un un caso feliz con foco en salida observable / trazabilidad "rica"

```python
df = make_happy_path_rich(
    filas=12,
    seed=42,
    extra_columns=[
        "household_id",
        "source_person_id",
        "activity_destination",
        "travel_time_min",
        "fare_amount",
    ],
    base_fields=[
        "mode",
        "purpose",
        "trip_weight",
        "user_gender",
        "origin_municipality",
        "destination_municipality",
    ],
)

display(df.head())
```

### 3) H3 derivable

```python
df = make_h3_derivable(filas=8, seed=42)

assert "origin_h3_index" not in df.columns
assert "destination_h3_index" not in df.columns

print(df.columns.tolist())
display(df.head())
```

Con override util: Cambiar el formato espacial, para no depender solo del default.

```python
df = make_h3_derivable(
    filas=10,
    seed=42,
    coord_format="dd_string",
    base_fields=["mode", "purpose"],
)

display(df.head())
```

### 4) Tier 2 válido

```python
df = make_tier2_valid(filas=8, seed=42)

display(df.head())
df[["origin_time_local_hhmm", "destination_time_local_hhmm"]].head()
```

Con override util: añadir segmentación temporal/categórica adicional.

```python
df = make_tier2_valid(
    filas=10,
    seed=42,
    base_fields=[
        "origin_time_local_hhmm",
        "destination_time_local_hhmm",
        "mode",
        "day_type",
        "time_period",
    ],
)

display(df.head())
```

### 5) Tier 2 mixto con inválidos

```python
df = make_tier2_mixed_invalid(filas=10, seed=42)

display(df.head())
df[["origin_time_local_hhmm", "destination_time_local_hhmm"]].head(10)
```

Versión más agresiva:

```python
df = make_tier2_mixed_invalid(
    filas=10,
    seed=42,
    mostly_invalid=True,
)
df[["origin_time_local_hhmm", "destination_time_local_hhmm"]].head(10)
```

Con override: Util para meter un poco de nulidad y tener HH:MM validos e invalidos

```python
df = make_tier2_mixed_invalid(
    filas=14,
    seed=42,
    null_ratio={
        "origin_time_local_hhmm": 0.20,
        "destination_time_local_hhmm": 0.30,
    },
    base_fields=["mode", "purpose"],
)

display(df.head(14))
```

### 6) Multistage válido

```python
df = make_multistage_valid(
    filas=15,
    seed=42,
    max_movements_per_trip=3,
)

display(df.head())
df[["trip_id", "movement_seq", "movement_id"]].head(15)
```

Con override: útil para representar viajes con varias filas por trip_id, y dialoga bien con la idea de movement vs trip del formato. Tambien es util para enriquecerlo con extras por etapa/viaje.

```python
df = make_multistage_valid(
    filas=18,
    seed=42,
    max_movements_per_trip=4,
    extra_columns=["stage_count", "travel_time_min", "public_transport_route_code"],
    base_fields=["mode", "purpose", "trip_weight"],
)

df.head(18)
```

### 7) Dominios extendidos

```python
df = make_extended_domains(
    filas=12,
    seed=42,
    include_noise=True,
)

display(df.head())
```

Con override: util para agregar algo de ruido y más de un campo extendido.

```python
df = make_extended_domains(
    filas=15,
    seed=42,
    extra_value_domains={
        "purpose": ["canon", "finos"],
        "mode": ["canon", "submodes"],
        "time_period": ["canon", "eod_finos"],
    },
    noise_ratio={
        "purpose": 0.3,
        "mode": 0.2,
    },
)

df.head(15)
```

### 8) Missing required

```python
df = make_missing_required(filas=6, seed=42)

display(df.head())
df.columns.tolist()
```

Indicando explícitamente qué required faltan:

```python
df = make_missing_required(
    filas=6,
    seed=42,
    missing_fields=["user_id", "origin_latitude"],
)
display(df.head())
```

Con override: util para quitar requeridos de manera aleatoria

```python
df = make_missing_required(
    filas=8,
    seed=42,
    n_random_missing_required=2,
)
display(df.head())
```

### 9) Duplicate movement_id

```python
df = make_duplicate_movement_id(filas=8, seed=42)

display(df.head())
df["movement_id"].value_counts().head()
```

Duplicando filas completas:

```python
df = make_duplicate_movement_id(
    filas=8,
    seed=42,
    full_rows=True,
)
display(df[3:8].head())
df.duplicated()
```

Con override: útil para volverlo un poco más rico para que no sea solo una tabla mínima con IDs repetidos

```python
df = make_duplicate_movement_id(
    filas=10,
    seed=42,
    base_fields=["mode", "purpose", "trip_weight"],
    extra_columns=["household_id", "travel_time_min"],
)

df.head(10)
```
