### 1) Happy path mínimo canónico

### 2) Happy path con algunos campos base canónicos

Ahora agregamos un poco de semántica útil, pero todavía sin rarezas.

### 3) Happy path con correspondencia de nombres

Este caso sirve para probar datasets fuente con nombres no canónicos, pero todavía válidos para import usando field_correspondence.

```python
df = generate_synthetic_trip_dataframe(
    filas=6,
    seed=1,
    field_correspondence={
        "user_id": "viajero_id",
        "origin_latitude": "o_lat",
        "origin_longitude": "o_lon",
    },
)

print(df.head())
print(df.columns.tolist())
```

```python
display(df)
```

### 4) Happy path Tier 2 con HH:MM válidos

Aquí hacemos un caso limpio, pero usando temporalidad Tier 2.

```python
df = generate_synthetic_trip_dataframe(
    filas=8,
    seed=3,
    tier_temporal="tier_2",
    tier2_hhmm_format="valid_hhmm",
    base_fields=[
        "origin_time_local_hhmm",
        "destination_time_local_hhmm",
        "mode",
    ],
)

print(df[
    [
        "movement_id",
        "origin_time_local_hhmm",
        "destination_time_local_hhmm",
        "mode",
    ]
].head())
```

```python
display(df)
```

### 5) Happy path con H3 derivable

Este es importante porque se parece mucho a un caso de integración real: el dataset no trae H3, pero sí trae coordenadas válidas y el import podría derivarlos.

Para este caso hay que combinar correctamente los argumentos. No basta con poner h3_mode="omitted_derivable": además hay que omitir explícitamente las columnas H3.

```python
df = generate_synthetic_trip_dataframe(
    filas=10,
    seed=4,
    omit_required_fields=["origin_h3_index", "destination_h3_index"],
    h3_mode="omitted_derivable",
)

print(df.head())
print(df.columns.tolist())
```

```python
display(df)
```

### 6) Happy path multistage

Ahora subimos complejidad estructural: varios movements pueden compartir trip_id.

```python
df = generate_synthetic_trip_dataframe(
    filas=12,
    seed=11,
    trip_structure="multistage",
    max_movements_per_trip=3,
    base_fields=["mode", "purpose", "trip_weight"],
)

print(df[["trip_id", "movement_seq", "movement_id", "mode", "purpose"]].head(12))
print(df["trip_id"].value_counts().head())
```

```python
display(df)
```

### 7) Dominios extendidos controlados

Aquí probamos un dataset aún válido, pero ya más interesante: usamos un campo categórico con mezcla de valores canónicos y extendidos.

```python
df = generate_synthetic_trip_dataframe(
    filas=10,
    seed=5,
    base_fields=["purpose", "user_gender", "day_type"],
    extra_value_domains={
        "purpose": ["canon", "finos"]
    },
    categorical_sampling_policy={
        "purpose": {"extended_ratio": 0.4}
    },
)

print(df[["purpose", "user_gender", "day_type"]].head(10))
print(df["purpose"].value_counts())
```

```python
display(df)
```

### 8) Dataset mas "rico": base + extras + nulos controlados

Este ya se parece mas a un caso de integración natural

```python
df = generate_synthetic_trip_dataframe(
    filas=15,
    seed=9,
    base_fields=["mode", "purpose", "trip_weight", "user_gender"],
    extra_columns=["household_id", "activity_destination", "travel_time_min"],
    null_ratio={
        "trip_weight": 0.2,
        "activity_destination": 0.3,
    },
)

print(df.head())
print(df.isna().mean().sort_values(ascending=False).head(10))
```

```python
display(df)
```

### 12) Ejemplo más complejo para parecerse más a una fuente real

Hacemos una configuración compleja pero todavía plausible/natural, pensada para parecerse más a una fuente real “rica”, no completamente limpia, pero tampoco destruida.

```python

df = generate_synthetic_trip_dataframe(
    filas=120,
    seed=20260322,

    # Estructura general
    duplicate_mode="none",
    trip_structure="multistage",
    max_movements_per_trip=3,

    # Temporalidad y espacialidad
    tier_temporal="tier_1",
    tier1_datetime_format="offset_string",
    coord_format="dms",
    h3_mode="provided_valid",

    # Campos base/opcionales Golondrina
    base_fields=[
        "mode",
        "purpose",
        "day_type",
        "time_period",
        "user_gender",
        "user_age_group",
        "income_quintile",
        "trip_weight",
        "origin_municipality",
        "destination_municipality",
    ],

    # Dominios extendidos
    extra_value_domains={
        "mode": ["canon", "submodes"],
        "purpose": ["canon", "finos"],
        "time_period": ["canon", "eod_finos"],
        "user_gender": ["canon", "es"],
        "income_quintile": ["canon", "tramos"],
    },

    # Cómo se muestrean esos dominios
    categorical_sampling_policy={
        "mode": {"canonical_ratio": 0.75, "ensure_presence": True},
        "purpose": {"canonical_ratio": 0.70, "ensure_presence": True},
        "time_period": {"canonical_ratio": 0.80},
        "user_gender": {"canonical_ratio": 0.90},
        "income_quintile": {"canonical_ratio": 0.85},
    },

    # Extras
    extra_columns=[
        "household_id",
        "source_person_id",
        "stage_count",
        "activity_destination",
        "travel_time_min",
        "fare_amount",
        "public_transport_route_code",
        "origin_venue_category",
        "destination_venue_category",
        "household_income_clp",
        "education_level",
        "activity_status",
        "driver_license_class",
    ],

    # Nulos realistas en algunas columnas
    null_ratio={
        "trip_weight": 0.08,
        "activity_destination": 0.12,
        "fare_amount": 0.10,
        "destination_venue_category": 0.18,
        "household_income_clp": 0.15,
        "education_level": 0.06,
    },

    # Patrones acoplados de incompletitud
    paired_missingness={
        "destination_incomplete": 0.05,
    },

    # Algo de contaminación categórica
    noise_ratio={
        "purpose": 0.04,
        "mode": 0.03,
        "user_gender": 0.02,
    },

    # Algunas columnas con tipo semánticamente "malo"
    type_corruption={
        "trip_weight": {"mode": "numeric_as_string", "ratio": 0.08},
        "household_income_clp": {"mode": "numeric_as_string", "ratio": 0.12},
        "fare_amount": {"mode": "stringified", "ratio": 0.10},
    },
)
```

Guardamos el dataset para una mejor inspeccion

```python
df.to_csv(DATA_PATH / "test_dataset.csv", sep=",", index=False)
```
